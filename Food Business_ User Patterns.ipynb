### country with maximum transcations on zomato


```python
#using a pie chart for vissualization

country_names = country_names[:len(country_val)]

plt.pie(country_val, labels=country_names)
plt.show()

```

## co-relations

```python
rating_info = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns ={0: 'Rating count'})
rating_info
```

### observations part 2

```python
#Majority of the people haven't given any ratings ever. 
#following diagram explains the corelation between number of ratings actually given ( the rating count)
#and its impact on aggregate. 
```

```python
sns.barplot(x="Aggregate rating",y= "Rating count", data = rating_info)
```

```python
#to enhance the visualization 
import matplotlib
matplotlib.rcParams['figure.figsize']= (12,7)
sns.barplot(x="Aggregate rating",y= "Rating count", hue = "Rating color", data = rating_info)
```

```python
# mapping the colors according to the provided data for better understanding

sns.barplot(x="Aggregate rating",y= "Rating count", hue = "Rating color", data = rating_info, palette = ['white', 'red','orange','yellow','green','green'])
```
